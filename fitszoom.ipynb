```python
from zoom import FITSZoom

# Load a sample FITS image (M51, the Whirlpool Galaxy, VLA radio data)
fz = FITSZoom('images/whirlpool_vla.fits')

# Zoom in on the galaxy center (RA, Dec in degrees) with a 300x300 pixel cutout
cutout = fz.zoom(coordinates=(202.469575, 47.195258), cutout_dimensions=(300, 300))

print('Original image shape:', fz.data.shape)
print('Cutout shape:', cutout.data.shape)

# Visualize the cutout
fig, ax = plt.subplots(subplot_kw={'projection': cutout.wcs})
norm = ImageNormalize(cutout.data, interval=ZScaleInterval())
ax.imshow(cutout.data, origin='lower', cmap='inferno', norm=norm)
ax.set_xlabel('RA')
ax.set_ylabel('Dec')
ax.set_title('Zoomed cutout: M51 (VLA)')
plt.show()

# Save the cutout to a new FITS file
fz.save('images/whirlpool_vla_cutout.fits')
```
